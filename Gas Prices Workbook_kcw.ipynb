```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file= "PET_PRI_GND_DCUS_NUS_W_csv.csv"
data_gas_price_df = pd.read_csv(csv_file)
data_gas_price_df.head()
```

```python
header_row = 1
data_gas_price_df.columns = data_gas_price_df.iloc[header_row]
data_gas_price_df.head()
```

```python
data_gas_price_df = data_gas_price_df.drop(header_row)
header_fixed_df = data_gas_price_df.reset_index(drop=True)
header_fixed_df.head()
```

```python
sourcekey_drop= header_fixed_df.drop(header_fixed_df.index[0])
sourcekey_drop.head()
```

```python
sourcekey_drop["Date"] =  pd.to_datetime(sourcekey_drop["Date"])
sourcekey_drop.head()
```

```python
start_date = "1994-11-28"
end_date= "2021-04-26"

after_start_date = sourcekey_drop["Date"] >= start_date
before_end_date = sourcekey_drop["Date"] <= end_date
between_two_dates = after_start_date & before_end_date
clean_dates_df = sourcekey_drop.loc[between_two_dates]
clean_dates_df
```

```python
colu = {}
for col in clean_dates_df:
    colu[col] = col.replace('Weekly U.S. ','').replace(' Retail Gasoline Prices  (Dollars per Gallon)','').replace(' Retail Prices  (Dollars per Gallon)','').replace(' ','_').lower()
colu
    
```

```python
clo_rename= clean_dates_df.rename(columns = colu)
clo_rename.head()
```

```python
c_df= clo_rename.drop('no_2_diesel_ultra_low_sulfur_(0-15_ppm)', axis=1)
cleaned_df= c_df.drop('no_2_diesel_low_sulfur_(15-500_ppm)', axis=1)
cleaned_df
```

```python
cleaned_df.to_csv('./gas_prices_cleaned.csv',index=False, header= True)
```
